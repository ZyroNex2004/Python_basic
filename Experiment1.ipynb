```python
# CHECK THAT A GIVEN NUMBER IF EVEN OR ODD
a=int(input("ENETER THE NUMBER: "))
if a%2==0:
    print("Even number : ",a)

else:
    print("Odd number : ",a)
```

```python
# LARGEST AMONG 2 NUMBER
a=int(input("ENETER THE FIRST NUMBER: "))
b=int(input("ENETER THE Second NUMBER: "))
if a>b:
    print("Largest number is 1st: ",a)

else:
    print("Largest number is 2nd: ",b)
```

```python
# check the given number is prime or not
a=int(input("ENETER THE FIRST NUMBER: "))
if a<=1:
    print("Non prime : ",a)

else:
    for i in range (2,a+1):
        if(a%i==0):
            print("Non prime : ",a)
            break

    else:
        print("Prime : ",a)
```

```python
# Find the CI and take the value from the user.

Principal=int(input("ENETER THE Principal Amount: "))
Time=int(input("ENETER THE Time: "))
Rate=int(input("ENETER THE Rate: "))
result=Principal*((1+(Rate/100))**Time)
print("Compound Interest : ",result)
```
